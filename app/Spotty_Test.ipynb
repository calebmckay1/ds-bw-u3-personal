```python
# to get access to the search function

# in a seperate file called ".env" put 2 variables:
# client_id="get this from spotify api dev dashboard"
# client_secret="get this from spotify api dev dashboard too"

# then run all cells 


import os
from dotenv import load_dotenv

load_dotenv()

client_id = os.getenv("client_id")
client_secret = os.getenv("client_secret", default="super secret")
```

```python
from spotify_client import SpotifyAPI
```

```python
spotify = SpotifyAPI(client_id, client_secret)
```

```python
# .search(query, search_type=)
# search_type= can be artist, track, album.
# and query is the term you are searching for.
# default search type is track.

spotify.search("ERASE U", search_type='track')
```
